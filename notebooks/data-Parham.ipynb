```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
## Read the data from folder data called coffee-location.csv
df = pd.read_csv('../data/coffee-location.csv')
```

```python
df.head()
```

```python
## fixing repeated rows for some of the countries due to "Price_grower" column 
df.groupby(['Country','Year']).mean().reset_index() 
```

```python
## deleting the repeated rows for each country and year combination
df = df.drop_duplicates(subset=['Country','Year'], keep='first')
```

```python
## turning the colmn 'Unnamed: 0' into a column called 'row_number'
df = df.rename(columns={'Unnamed: 0': 'row_number'})
```

```python
## making the row_number column the index of the dataframe
df = df.set_index('row_number')
```

```python
## making the 'row_number' startin from 0 with a step of 1
df.index = np.arange(0, len(df))
```
