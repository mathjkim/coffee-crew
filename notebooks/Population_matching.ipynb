Loading the required packages

```python
import numpy as np
import pandas as pd
```

Loading the data. The coffee_data contains the location of the coffee producers, consumers and imports and exports. The population data contains the population of different countries. We extract the 30 years we are interested in. countries contain the list of the countries from the population dataset. The countries in the coffee data is a subset  of the countries contained the population dataset.

```python
coffee_data = pd.read_csv('./data/coffee-location-parham.csv')
population_data =pd.read_csv("./Population_trend.csv", skiprows=4,usecols=[x for x in range (34,65)])
countries = pd.read_csv("./Population_trend.csv", skiprows=4,usecols=['Country Name'])
countries.head()
```

```python
coffee_data.head()
```

```python
i=0
s=0
missing_val = 0 # just an index to locate the positions, 
#if there is a discrepency between the names of the countries in two data sets
pop = [] # array containing the population of the countries in the same order as 
#that in the coffe data set

while i<len(coffee_data["Country"]):
    missing_val = 0
    if(i>0 and (coffee_data["Country"][i]==coffee_data["Country"][i-1])):
        if(np.isnan(pop[-1])):
            pop.append(np.nan)
            i=i+1
            print(i)
            missing_val = 1
           
        else:
            pop.append(population_data[str(coffee_data["Year"][i])][s])
            i=i+1
        
        
    for j in range(len(countries["Country Name"])):
        if(coffee_data["Country"][i]==countries["Country Name"][j]):
            pop.append(population_data[str(coffee_data["Year"][i])][j])
            i = i+1
            s = j
            missing_val = 1
            

            
    if(missing_val == 0):
        pop.append(np.nan)
        i=i+1
        print(i)
```

The corresponding population data in now added to the coffee data.

```python
coffee_data.replace(0,np.nan,inplace=True)
coffee_data["Population"] = np.array(pop).tolist()
```
